```python
import pandas as pd
import ast
from random import randint
df = pd.read_csv("dataset.csv")
```

```python
df.info()
```

```python
df_2 = df[["productId", "title", "description", "uom", "pricing.price"]]
```

```python
def extract_image(row):
    try:
        image = ast.literal_eval(row["productImage"])[0]["imageUrl"]
    except:
        image = None
    return image

df_2["Image"] = df.apply(extract_image, axis=1)
df_2
```

```python
df_2.to_csv("datasets/master.csv")
```

```python
stores = ["Walmart","Zehrs","Freshco","Food Basics","Canadian Superstore","Farahs","No Frills"]
```

```python


# Create a copy of df_2 to avoid modifying the original DataFrame
df_copy = df_2.copy()

for store in stores:
    discount_rate = randint(5, 20)
    # Apply discount to the pricing.price column and add the new column
    df_copy[store] = round(df_copy["pricing.price"] * (1 - discount_rate / 100), 2)

# Save the updated DataFrame to a CSV file
df_copy.to_csv("datasets/globaldataset.csv", index=False)
```
